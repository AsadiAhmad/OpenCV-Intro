# Practice 2:

Showing Image in GrayScale

```python
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread("Ahmad.jpg", cv2.IMREAD_COLOR)
cv2_imshow(img)
height, width, channels = img.shape

print(f"Image width M: {width} pixels")
print(f"Image height N: {height} pixels")
print(f"our image is : {width} * {height}")

gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

#Practice 3

Print image matrix

```python
for row in gray_image:
    print("[", end="")
    for pixel in row:
        print(str(pixel) + ",", end="")
    print()
```

#Practice 4

size of matrix is size of image

```python
m = 0
n = 0
for row in gray_image:
    n += 1
    for pixel in row:
        m += 1
print(f"size of m is : {m/n}")
print(f"size of n is : {n}")
```

#Practice 5

Mean of all pixels in grayscale image

```python
sigma = 0
for row in gray_image:
    for pixel in row:
        sigma += pixel

count = width * height
mean = sigma/count
print(f"Mean of image pixel is : {mean}")
```

#Practice 6

Calculating Variance

```python
sigma2 = 0
for row in gray_image:
    for pixel in row:
        sigma2 += (pixel - mean)*(pixel - mean)

variance = sigma2/count

print(f"Variance of image is : {variance}")
```
